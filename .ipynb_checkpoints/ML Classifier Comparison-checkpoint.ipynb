# Divide datasets into 2

```python
diabetesData['Age'].median()
```

```python
# divide the database into 2 such that 2 classifiers can be generated on different datasets
youngerDiabetes = diabetesData[diabetesData['Age'] < 29]
olderDiabetes = diabetesData[diabetesData['Age'] >= 29]
```

### Dataset A and Dataset B
- Split dataset A for training and testing
- Split dataset B for training and testing
- Train Tree A based on training data A
- Train Tree B based on training data B
- Test both trees based on testing data A + testing data B (combination/union)

# Create and Train 2 decision tree classifiers

```python
# select features (all columns except 'outcome')
features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

# data of X (predictors)
younger_X = youngerDiabetes[features]
older_X = olderDiabetes[features]

# data of Y (target)
younger_y = youngerDiabetes.Outcome
older_y = olderDiabetes.Outcome
```

```python
# splitting into validation(testing) and training
# the default test_size is 0.25 given not specifying both train_size and test_size (train_size = 0.75
# and 25:75 is good rool of thumb
# Regarding 'stratify' - It makes a split so that the proportion of values in the sample produced
# will be the same as the proportion of values provided to the parameter stratify
younger_X_train, younger_X_test, younger_y_train, younger_y_test = train_test_split(younger_X, younger_y, random_state=1, shuffle = True, stratify=younger_y)
older_X_train, older_X_test, older_y_train, older_y_test = train_test_split(older_X, older_y, random_state=1, shuffle = True, stratify=older_y)
```

```python
# specify decision tree model
younger_decisionTree_model = DecisionTreeClassifier(random_state=1, max_depth=6) # explore this later to justify why depth=10 (could be 5? or smaller)
older_decisionTree_model = DecisionTreeClassifier(random_state=1, max_depth=6)

# train(fit) decision tree classifier
younger_decisionTree_model.fit(younger_X_train, younger_y_train)
older_decisionTree_model.fit(older_X_train, older_y_train)
```

```python
# create test cases to examine the performance of classifiers
testcases = pd.DataFrame(older_X_test.append(younger_X_test))
testoutcome = pd.DataFrame(older_y_test.append(younger_y_test))
```

# Print accuracies and precisions of classifiers
## Improve these accuracies to make more realistic! (explore this later)

### Accuracy = a ratio of correctly predicted observation to the total observations
- A great measure but only when you have symmetric datasets where values of false positive and false negatives are almost same
- Accuracy = TP+TN/TP+FP+FN+TN

### Precision = a ratio of correctly predicted positive observations to the total predicted positive observations
- High precision relates to the low false positive rate
- Precision = TP/TP+FP

-- Source: https://blog.exsilio.com/all/accuracy-precision-recall-f1-score-interpretation-of-performance-measures/

```python
# predict the response for test dataset
older_y_pred = older_decisionTree_model.predict(older_X_test)
older_y_pred_opposite = older_decisionTree_model.predict(younger_X_test)
#older_y_pred_total = older_decisionTree_model.predict(testcases)
younger_y_pred = younger_decisionTree_model.predict(younger_X_test)
younger_y_pred_opposite = younger_decisionTree_model.predict(older_X_test)
#younger_y_pred_total = younger_decisionTree_model.predict(testcases)

print("Accuracy of 'younger' decision tree against 'younger' test:", accuracy_score(younger_y_test, younger_y_pred))
print("Accuracy of 'younger' decision tree against 'older' test:", accuracy_score(older_y_test, younger_y_pred_opposite))
print("Accuracy of 'older' decision tree against 'older' test:", accuracy_score(older_y_test, older_y_pred))
print("Accuracy of 'older' decision tree against 'younger' test:", accuracy_score(younger_y_test, older_y_pred_opposite))

#print("Accuracy of younger decision tree against all test cases:", accuracy_score(testoutcome, younger_y_pred))
#print("Accuracy of older decision tree against all test cases:", accuracy_score(testoutcome, older_y_pred))

print("\n")
print("How similar the results of older tree and younger tree tested against 'younger' testcases are:", accuracy_score(older_y_pred_opposite, younger_y_pred)) # tells me how those 2 classifiers are similar to each other
# given that this was 0.7934782608695652 where the test cases for younger_X are 96, approx 20 cases disagreed
print("How similar the results of older tree and younger tree tested against 'older' testcases are:", accuracy_score(younger_y_pred_opposite, older_y_pred))
print("-- These tell me how similar the 2 trained decision classifiers are")
print("-- There are %d test cases in younger, and %d test cases in older" %(len(younger_X_test), len(older_X_test)))
print("-- Given that they had approximately 79 percent of similarity over 'younger' testcases, around %d testcases were disagreed" %(round(len(younger_X_test) * 0.21)))
print("-- Given that they had approximately 76 percent of similarity over 'older' testcases, around %d testcases were disagreed" %(round(len(older_X_test) * 0.24)))
```

```python
print("Average precision:", average_precision_score(younger_y_test, younger_y_pred))
print("Average precision:", average_precision_score(older_y_test, older_y_pred))

#print("Average precision:", average_precision_score(testoutcome, younger_y_pred))
#print("Average precision:", average_precision_score(testoutcome, older_y_pred))
```

# Optimising decision tree performance
### Decision Tree Criterion
- {“gini”, “entropy”}, default=”gini”
- Both gini and entropy options are measures of 'impurity' of a node. A node having multiple classes is impure while a node having only one class is pure (most leaves are pure unless the tree has no attributes to split further on)
### Decision Tree Splitter
- {“best”, “random”}, default=”best”
### Decision Tree Max Depth
- int or None, optional (default = None) 

# WHY did they disagreed on those cases?
- Use feature_importance
- Train a new classifier which has new labels e.g. train on the same dataset but use for comparison but the labels will be whether they disagree or not...
- The datasets we used for comparison was younger_X_test
- Whether they disagree or not (the labels = np.logical_xor(older_y_pred_opposite, younger_y_pred)) this is the TARGET
- Don't train/test split just use all of them for training!! 
- For this ML classifier, print feature_importance_ (high importance tells me why they disagreed)
- No decision tree? Use logistic regression...

# Print feature_importance_ of younger testcases classifier
- This tells me which features effected how significantly on making disagreement between classifiers

```python
# Computing the disagreed cases (i.e. either when A right but B wrong / A wrong but B right)
younger_disagreed_testcases = younger_X_test[np.logical_xor(older_y_pred_opposite, younger_y_pred)]
younger_agreed_testcases = younger_X_test.loc[younger_X_test.index.difference(younger_disagreed_testcases.index)]

younger_disagreed_testcases['Agreed'] = 0
younger_agreed_testcases['Agreed'] = 1

all_younger_testcases = pd.DataFrame(younger_disagreed_testcases.append(younger_agreed_testcases))

features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

# don't test this ML classifier
# data of X (predictors)
younger_testcases_X_train = all_younger_testcases[features]

# data of Y (target)
younger_testcases_y_train = all_younger_testcases.Agreed

younger_testcases_decisionTree_model = DecisionTreeClassifier(random_state=1) # explore this later to justify why depth=10 (could be 5? or smaller)

# train(fit) decision tree classifier
younger_testcases_decisionTree_model.fit(younger_testcases_X_train, younger_testcases_y_train)

younger_testcases_feature_importance = younger_testcases_decisionTree_model.feature_importances_

younger_testcases_feature_importance_output = pd.DataFrame(younger_testcases_feature_importance, features)
younger_testcases_feature_importance_output.set_axis(['Output'], axis=1, inplace=True)
print(younger_testcases_feature_importance_output)
ax = younger_testcases_feature_importance_output.plot.barh()
```

```python
def return_disagreed_cases(testcases, prediction_result1, prediction_result2):
    filters = np.logical_xor(prediction_result1, prediction_result2)
    disagreed = testcases[filters]
    disagreed['Agreed'] = 0
    return disagreed

def return_agreed_cases(testcases, prediction_result1, prediction_result2):
    filters = return_disagreed_cases(testcases, prediction_result1, prediction_result2)
    agreed = testcases.loc[testcases.index.difference(filters.index)]
    agreed['Agreed'] = 1
    return agreed

#train_decision_tree(train_X, train_y, model)
'''def create_(agreed, disagreed, features):
    all_cases = pd.DataFrame(agreed.append(disagreed))
    train_X = all_cases[featuers]
    train_y = all_cases.Agreed
    model = create_decision_tree()'''
```

# Print feature_importance_ of older testcases classifier

```python
# Computing the disagreed cases (i.e. either when A right but B wrong / A wrong but B right)
older_disagreed_testcases = older_X_test[np.logical_xor(younger_y_pred_opposite, older_y_pred)]
older_agreed_testcases = older_X_test.loc[older_X_test.index.difference(older_disagreed_testcases.index)]

older_disagreed_testcases['Agreed'] = 0
older_agreed_testcases['Agreed'] = 1

all_older_testcases = pd.DataFrame(older_disagreed_testcases.append(older_agreed_testcases))

features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

# don't test this ML classifier
# data of X (predictors)
older_testcases_X_train = all_older_testcases[features]

# data of Y (target)
older_testcases_y_train = all_older_testcases.Agreed

older_testcases_decisionTree_model = DecisionTreeClassifier(random_state=1) # explore this later to justify why depth=10 (could be 5? or smaller)

# train(fit) decision tree classifier
older_testcases_decisionTree_model.fit(older_testcases_X_train, older_testcases_y_train)

older_testcases_feature_importance = older_testcases_decisionTree_model.feature_importances_

older_testcases_feature_importance_output = pd.DataFrame(older_testcases_feature_importance, features)
older_testcases_feature_importance_output.set_axis(['Output'], axis=1, inplace=True)
print(older_testcases_feature_importance_output)
ax = older_testcases_feature_importance_output.plot.barh()
```
